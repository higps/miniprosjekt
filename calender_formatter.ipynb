```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("calender.csv")
df_draft = df.copy()
```

```python
df_draft["date"] = pd.to_datetime(df_draft["date"]).apply(lambda x: x.strftime("%d.%m.%Y"))
```

```python
df_work_month=df_draft[df_draft["red_day_not_sun"] == 0].groupby("month_year_name", as_index=False)["red_day_not_sun"].count()
df_work_month
```

```python
df_work_month.to_csv("work_days.csv")
```

```python
df_draft.to_csv("formatted_calender.csv")
```
